```python
import importlib
from IPython.display import display

import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from pandas.tseries.offsets import Day, MonthEnd
from datetime import datetime
from datetime import timedelta
from datetime import date

from account import Index, Account, Merge
```

# Input Basic Data

```python
mtrt = 26 #개월
prd_prjt = 24 #개월(프로젝트 기간)
repay_rate = 1.0 # 상환비율(매출 발생금액 중 상환계좌 이체 비율)
idx = Index('2021-04', periods=12, freq='M')
```

# Input Sales Data

```python
# Initial Sales Data
# Product A : 상온창고
prdtA = Account(idx, "ProductA", ("Prdt", "A"))
prdtA.rent = 21_000 # 월 임대료 21,000원/평 가정
prdtA.area = 8_000 # 면적 8,000평
prdtA.cap = 0.053 # Cap rate 5.3%
prdtA.amt = prdtA.rent * 12 * prdtA.area / prdtA.cap / 1_000_000
prdtA.addscdd(prdtA.index[0], prdtA.amt)
prdtA.subscdd(prdtA.index[11], prdtA.amt)
prdtA.addamt(prdtA.index[0], prdtA.df.loc[prdtA.index[0], 'add_rsdl_cum'])
prdtA.subamt(prdtA.index[11], prdtA.df.loc[prdtA.index[11], 'sub_rsdl_cum'])

# Product B : 저온창고
prdtB = Account(idx, "ProductB", ("Prdt", "B"))
prdtB.rent = 50_000 # 월 임대료 50,000원/평 가정
prdtB.area = 6_500 # 면적 6,500평
prdtB.cap = 0.058 # Cap rate 5.8%
prdtB.amt = prdtB.rent * 12 * prdtB.area / prdtB.cap / 1_000_000
prdtB.addscdd(prdtB.index[0], prdtB.amt)
prdtB.subscdd(prdtB.index[11], prdtB.amt)
prdtB.addamt(prdtB.index[0], prdtB.df.loc[prdtB.index[0], 'add_rsdl_cum'])
prdtB.subamt(prdtB.index[11], prdtB.df.loc[prdtB.index[11], 'sub_rsdl_cum'])

# Merge
sales = Merge({'prdtA':prdtA, 'prdtB':prdtB})
```

# Input Cost Data

```python
dct = {}

# 공정률
idx.prcs_rate = Series(np.ones(len(idx)) / len(idx), 
                       index=idx.index)

# Initial Cost Data
cstlnd = Account(idx, "Cost_Land", ("Cost", "Land"))
cstlnd.amt = 15_500.000 # 최초 1회 지급
cstlnd.addscdd(idx.index[0], cstlnd.amt)
dct['cstlnd'] = cstlnd

cstcstrn = Account(idx, "Cost_Construction", ("Cost", "Construction"))
cstcstrn.amt = 50_000.000 # 공사비(지급) : 공정률에 따라 지급
cstcstrn.addscdd(idx.index, cstcstrn.amt * idx.prcs_rate)
dct['cstcstrn'] = cstcstrn

cstcstrnb = Account(idx, "Cost_Construction B", ("Cost", "ConstructionB"))
cstcstrnb.amt = 3_000.000 # 최초 1회 지급
cstcstrnb.addscdd(idx.index[0], cstcstrnb.amt)
dct['cstcstrnb'] = cstcstrnb

cstmrktg = Account(idx, "Cost_Marketing", ("Cost", "Marketing"))
cstmrktg.amt = 2_000.000 # 공정률에 따라 지급
cstmrktg.addscdd(idx.index, cstmrktg.amt * idx.prcs_rate)
dct['cstmrktg'] = cstmrktg

csttax = Account(idx, "Cost_Tax", ("Cost", "Tax"))
csttax.amt = 3_000.000 # 준공시 1회 지급
csttax.addscdd(idx.index[11], csttax.amt)
dct['csttax'] = csttax

cstoprtg = Account(idx, "Cost_Operating", ("Cost", "Operating"))
cstoprtg.amt = 2_200.000
cstoprtg.addscdd(idx.index, cstoprtg.amt * idx.prcs_rate)
dct['cstoprtg'] = cstoprtg

cost = Merge(dct)
```
